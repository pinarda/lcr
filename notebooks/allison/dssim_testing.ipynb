```python
####TS

import numpy as np
a_ts_ssim = np.zeros((10,10))
a_ts_ssim_fp_orig =np.zeros((10,10))
a_ts_ssim_fp = np.zeros((10,10))

zfp = ["p6", "p8", "p10", "p12", "p14", "p16", "p18", "p20", "p22", "p24"]


for j, val in enumerate(zfp):
    print(val)
    for i in range(10):
        ts_orig = all_ts_col["TS"].sel(collection="orig").isel(time=i).compute()
        ts_comp = all_ts_col["TS"].sel(collection=val).isel(time=i).compute()
        ts_differ = ldcpy.calcs.Diffcalcs(ts_orig, ts_comp)
        ts_ssim_fp = ts_differ.get_diff_calc('ssim_fp')
        ts_ssim = ts_differ.get_diff_calc('ssim')
        ts_ssim_fp_orig = ts_differ.get_diff_calc('ssim_fp_orig')

    


        a_ts_ssim[j,i] = ts_ssim
        a_ts_ssim_fp_orig[j,i] = ts_ssim_fp_orig
        a_ts_ssim_fp[j,i] =  ts_ssim_fp

```
