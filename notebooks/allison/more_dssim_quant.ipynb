```python
import numpy as np

vnames = ["TS", "PRECT", "PS", "FLUT", "QFLX"]
vnames = ["TS", "PRECT"]
ds = [all_ts_col, all_prect_col, all_ps_col, all_flut_col, all_qflx_col]
ds = [all_ts_col, all_prect_col]

val = "p14"
q_values = [32, 64, 128, 256, 512, 1024, 2048, 0]
q_values = [16, 32, 64, 128, 256, 512, 1024, 2048, 4096, 8192, 0]

num = len(q_values)
nvar = len(vnames)

a_dssim_exp = np.zeros((num, nvar))
ts_ssim = np.zeros(nvar)
ts_dssim = np.zeros(nvar)

for j, varname in enumerate(vnames):
    print(varname)
    #just doing t=0
    for i in range(1):
        ts_orig = ds[j][varname].sel(collection="orig").isel(time=i).compute()
        ts_comp = ds[j][varname].sel(collection=val).isel(time=i).compute()

        ts_differ = ldcpy.calcs.Diffcalcs(ts_orig, ts_comp, "cam-fv")
        
        ts_dssim[j] = ts_differ.get_diff_calc('ssim_fp')
        ts_ssim[j] = ts_differ.get_diff_calc('ssim')
    
        for q, q_val in enumerate(q_values):
            ts_differ.quant = q_val
            ts_differ.k1 = 1e-8
            ts_differ.k2 = 1e-8
            #print('----------------------')
            #print("quant = ", ts_differ.quant)
        
            ts_dssim_exp = ts_differ.get_diff_calc('ssim_fp_exp')
            #print(ts_dssim_exp)
            a_dssim_exp[q,j] = ts_dssim_exp
        

        
        
```

```python
my_labels = []
for q in q_values:
  my_labels.append(str(q)) 

nvar = len(vnames)

from matplotlib import pyplot as plt
#fig, axs = plt.subplots(nvar,1,figsize=(8,26))
fig, axs = plt.subplots(nvar,1,figsize=(8,10))

for p,name in enumerate(vnames):
    axs[p].plot(my_labels, a_dssim_exp[:,p] , 'r', label = "DSSIM-EXP")
    axs[p].plot(my_labels,  np.full(num, ts_ssim[p])    , 'g:',label = "Default SSIM")
    axs[p].plot(my_labels,  np.full(num, ts_dssim[p])    , 'r:',label = "Default DSSIM")

    #tmp_a = np.zeros(num*3)
    #np.concatenate((a_dssim[:,p],a_sf_ssim[:,p], a_ssim[:,p]), out = tmp_a)
    #minlim = min(tmp_a)
    #axs[p].set_ylim([minlim, 1.0])

    axs[p].legend(fontsize=14)
    axs[p].set_title(f"{name} - zfp14")
    axs[p].set_xlabel("quantization bins (0 = none)", fontsize=14)
    axs[p].set_ylabel("value of SSIM variant", fontsize=14)
```
